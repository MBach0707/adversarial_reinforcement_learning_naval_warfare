Script to determine the average distance between two points on a x by x grid. The distance is also known as the Manhattan or taxi distance.
Note that the optimal policy 

```python
import random as rn
import numpy as np
import tqdm as tqdm

size = [8, 8]

iterations = 10000000
manhattan_dist = []
euclidean_dist = []

for i in tqdm.tqdm(range(iterations)):
    x_loot = rn.randint(0, size[0])
    y_loot = rn.randint(0, size[1])
    x_pos = rn.randint(0, size[0])
    y_pos = rn.randint(0, size[1])
    x = abs(x_loot - x_pos)
    y = abs(y_loot - y_pos)
    manhattan_dist.append(x + y)
    euclidean_dist.append(np.sqrt(x**2 + y**2))

manhattan_avg = sum(manhattan_dist)/len(manhattan_dist)
euclidean_avg = sum(euclidean_dist)/len(euclidean_dist)
print(f"Manhattan distance: {manhattan_avg}")
print(f"Euclidean distance: {euclidean_avg}")
```
